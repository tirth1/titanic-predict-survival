```python
df = pd.concat([train, test], sort=False, ignore_index=True)
```

```python
df.isnull().mean().sort_values(ascending=False)
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

```python
# Checking correlations using Heatmap
fig, axs = plt.subplots(nrows=1, figsize=(13, 13))
sns.heatmap(df.corr(), annot=True, square=True, cmap='YlGnBu', linewidths=2, linecolor='black', annot_kws={'size':12})
```

```python
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.drop(['Ticket'], axis=1, inplace=True)
```

```python
df.drop(['Cabin'], axis=1, inplace=True)
```

```python
df.drop(['Name'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, drop_first = True)
```

```python
df.head()
```

```python
train = df[: len(train)]
```

```python
train.head()
```

```python
test = df[len(train):]
```

```python
test.head()
```

```python
train.head()
```

```python
y = train.Survived
X = train.drop('Survived', axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB 
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
```

```python
X = preprocessing.scale(X)
```

```python
print(cross_val_score(LogisticRegression(), X, y, cv=10, scoring='accuracy').mean())
```

```python
print(cross_val_score(SVC(), X, y, cv=10, scoring='accuracy').mean())
```

```python
print(cross_val_score(GaussianNB(), X, y, cv=10, scoring='accuracy').mean())
```

```python
print(cross_val_score(RandomForestClassifier(), X, y, cv=10, scoring='accuracy').mean())
```

```python
print(cross_val_score(DecisionTreeClassifier(), X, y, cv=10, scoring='accuracy').mean())
```

```python
print(cross_val_score(GradientBoostingClassifier(), X, y, cv=10, scoring='accuracy').mean())
```

```python
# Using train_test_split we split the data into train and validation data for testing
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
# final model
final_model = SVC(random_state=0)
```

```python
# Train final_model with train data
final_model.fit(X_train, y_train)
```

```python
# Predict final_model
predictions = final_model.predict(X_val)
```

```python
# Print out score
from sklearn.metrics import accuracy_score
print('Accuracy: ', accuracy_score(predictions, y_val))
```
